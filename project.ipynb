```python
#22 take a sample size 50 form column damage dealt for 100 times and calculate mean.plot it on histogram and comment on its distribution
l=[]
x=0
while x<100:
    sample=data["damageDealt"].sample(n=50).mean()
    sample=round(sample,2)
    l.append(sample)
    x+=1
print(l)
```

```python
%matplotlib inline
plt.figure(figsize=(10,8))
plt.hist(l,histtype='step');
plt.title("damage dealt by the player")
```
